# House Price Prediction

```python
%config IPCompleter.greedy = True
%matplotlib inline

# Pandas
import pandas as pd

# Numpy
import numpy as np

#Scipy
from scipy import stats

#Matplotlib
import matplotlib.pyplot as plt

# Sklearn
from sklearn.linear_model import LinearRegression, Ridge, Lasso, RidgeCV
from sklearn.model_selection import train_test_split, cross_val_score, cross_val_predict
from sklearn import metrics, svm
from sklearn.metrics import mean_squared_error, r2_score

# Seaborn
import seaborn as sns
sns.set(rc={'figure.figsize':(10,10)})
# sns.set(rc={'figure.figsize':(10,10)}, font_scale= 2)
```

## Load Training and Testing Data

```python
# Load training data

train_data = pd.read_csv("data/train.csv")
```

```python
# Load Testing Data

test_data = pd.read_csv("data/test.csv")
```

Removing Id Column From Both Train and Test Data

```python
# Remove Id Column from Training Data

train_data = train_data.drop(columns="Id")
```

```python
# Remove Id Column from Training Data

test_ids = test_data['Id']
test_data = test_data.drop(columns="Id")
```

 Defining Categorical To Numerical Functions

```python
## Categorical to Numerical conversion functions

def categorical_to_numerical_for_train(column):
    data, columns = train_data[column].factorize()
    return data

def categorical_to_numerical_for_test(column):
    data, columns = test_data[column].factorize()
    return data
```

```python
# List of Columns to Drop

to_drop = []
```

## Train Data Pre-Processing

```python
## pre-process training data

numerical_cols = train_data._get_numeric_data().columns.values
all_cols = train_data.columns.values
categorical_cols = np.setdiff1d(all_cols, numerical_cols)
```

Dealing With Numerical Data

```python
## check missing numerical data
numerical_na = train_data._get_numeric_data().isnull().mean(axis=0) # For Numerical Values
numerical_na = numerical_na.drop(numerical_na[numerical_na == 0].index).sort_values(ascending=False)
```

```python
## visualize missing data

f, ax = plt.subplots(figsize=(6, 3))
plt.xticks(rotation='90')
sns.barplot(x = numerical_na.index, y = 100*numerical_na)
plt.xlabel('Features', fontsize=15)
plt.ylabel('Percentage of missing values', fontsize=15)
plt.title('Percentage of missing data of features', fontsize=15)
```

```python
# set missing values to 0
for col in numerical_na.index.values:
    if (numerical_na[col] > 0.6):
        to_drop.append(col)
    train_data[col] = train_data[col].transform(lambda x: x.fillna(x.median()))
```

Dealing With Categorical Data

```python
## check missing categorical data
categorical_na = train_data[categorical_cols].isnull().mean(axis=0) # For Categorical Values
categorical_na = categorical_na.drop(categorical_na[categorical_na == 0].index).sort_values(ascending=False)
```

```python
# visualize missing data

f, ax = plt.subplots(figsize=(6,3))
plt.xticks(rotation='90')
sns.barplot(x = categorical_na.index, y = 100*categorical_na)
plt.xlabel('Features', fontsize=15)
plt.ylabel('Percentage of missing values', fontsize=15)
plt.title('Percentage of missing data of features', fontsize=15)
```

```python
# set missing values to Mode

for col in categorical_na.index.values:        
    if (categorical_na[col] > 0.6):
        to_drop.append(col)
    train_data[col] = train_data[col].transform(lambda x: x.fillna(x.mode()))
```

```python
# set categorical to numeric

for col in categorical_cols:
    train_data[col] = categorical_to_numerical_for_train(col)
```

## Test Data Pre-Processing

```python
## pre-process testing data

numerical_cols = test_data._get_numeric_data().columns.values
all_cols = test_data.columns.values
categorical_cols = np.setdiff1d(all_cols, numerical_cols)
```

Dealing With Numerical Data

```python
## check missing numerical data

numerical_na = test_data._get_numeric_data().isnull().mean(axis=0) # For Numerical Values
numerical_na = numerical_na.drop(numerical_na[numerical_na == 0].index).sort_values(ascending=False)
```

```python
## visualize missing data

f, ax = plt.subplots(figsize=(6,3))
plt.xticks(rotation='90')
sns.barplot(x = numerical_na.index, y = 100*numerical_na)
plt.xlabel('Features', fontsize=15)
plt.ylabel('Percentage of missing values', fontsize=15)
plt.title('Percentage of missing data of features', fontsize=15)
```

```python
# set missing values to 0

for col in numerical_na.index.values:
    test_data[col] = test_data[col].transform(lambda x: x.fillna(x.median()))
#     test_data[col] = test_data[col].transform(lambda x: x.fillna(x.median()))
```

Dealing With Categorical Data

```python
## check missing categorical data

categorical_na = test_data[categorical_cols].isnull().mean(axis=0) # For Categorical Values
categorical_na = categorical_na.drop(categorical_na[categorical_na == 0].index).sort_values(ascending=False)
```

```python
## visualize missing data

f, ax = plt.subplots(figsize=(6,3))
plt.xticks(rotation='90')
sns.barplot(x = categorical_na.index, y = 100*categorical_na)
plt.xlabel('Features', fontsize=15)
plt.ylabel('Percentage of missing values', fontsize=15)
plt.title('Percentage of missing data of features', fontsize=15)
```

```python
# set missing values to Mode

for col in categorical_na.index.values:
    test_data[col] = test_data[col].transform(lambda x: x.fillna(x.mode()))
```

```python
# set categorical to numeric

for col in categorical_cols:
    test_data[col] = categorical_to_numerical_for_test(col)
```

# Correlation of Features

Correlation Matrix

```python
plt.figure(figsize=(20,20))
sns.heatmap(test_data.corr(), vmax=1, cmap="Blues", square=True)
```

Columns Highly Correlated To Each Other

```python
# Create correlation matrix
corr_matrix = train_data.corr().abs()

# Select upper triangle of correlation matrix
upper = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(np.bool))

# Find index of feature columns with correlation greater than 0.8
high_corr = [column for column in upper.columns if any(upper[column] > 0.80)]

to_drop = high_corr + to_drop
```

Columns Highly Corrlated To SalePrice

```python
correlated_elements = train_data.corr()['SalePrice'].sort_values(ascending=False)

plt.figure(figsize=(20,20))
objects = correlated_elements.index.values
y_pos = np.arange(len(objects))
performance = correlated_elements.values

plt.barh(y_pos, performance, align='center', alpha=0.5)
plt.yticks(y_pos, objects)
plt.xlabel('Correlation')
plt.title('Correlation of features with SalePrice')

plt.show()
```

Deleting Least Correlated Columns To SalePrice

```python
for col in train_data.columns.values:
    if(correlated_elements[col] < 0.01 and correlated_elements[col] > -0.01):
        to_drop.append(col)  
```

```python
for col in train_data.columns.values:
    if col not in to_drop:
        val = train_data.groupby([col]).size()[train_data.groupby([col]).size().sort_values(ascending=False).index.values[0]]
        total = train_data.count().mode()[0]
        if((val/total) > 0.95):
            to_drop.append(col)
```

# Skewness And Kurtosis

```python
# function definition for checking skewness and kurtosis

def check_skew_kurt():
    #Check the distribution 
    sns.distplot(train_data['SalePrice'] , fit=stats.norm);
    (mu, sigma) = stats.norm.fit(train_data['SalePrice'])

    # mean, std, skewness and kurtosis
    print('mu = {:.2f} and sigma = {:.2f}'.format(mu, sigma))
    print("Skewness: %f" % train_data['SalePrice'].skew())
    print("Kurtosis: %f" % train_data['SalePrice'].kurt())
```

```python
# Check skewness and kurtosis for SalePrice

plt.figure(figsize=(5,4))
check_skew_kurt()
```

Applying Log To SalePrice To Reduce The Skewness And Kurtosis

```python
train_data["SalePrice"] = np.log(train_data["SalePrice"])
```

```python
# Check skewness and kurtosis for SalePrice after applying log

plt.figure(figsize=(5,4))
check_skew_kurt()
```

# Separating Prediction Value

```python
# Variable to Predict
result = train_data['SalePrice']
```

```python
# Removing SalePrice
train_data = train_data.drop(columns="SalePrice")
```

# Splitting Data Into Train And Test Set

```python
# Dividing the Train data into Train and Test

X_train, X_test, y_train, y_test = train_test_split(train_data, result, test_size=0.3, shuffle=True, random_state=1)
```

# Linear Regression

```python
# LINEAR REGRESSION
regressor = LinearRegression(fit_intercept=False, normalize=True)
regressor.fit(X_train, y_train)
linear_score = regressor.score(X_test, y_test)
print("Linear Score : ", linear_score)
linear_predictions = regressor.predict(X_test)

# plt.plot(np.e**housing_predictions - 1, np.e**y_test, 'ro')
# plt.show()

plt.figure(figsize=(10,8))
# plt.plot(y_test, linear_predictions, 'ro')
sns.regplot(x=np.exp(y_test), y=np.exp(linear_predictions))
plt.xlabel('Actual Value')
plt.ylabel('Prediction Value')
plt.show()

linear_rmse = np.sqrt(mean_squared_error(y_test, linear_predictions))
print("RMSE :", linear_rmse)
print("Linear Intercept", regressor.intercept_)
```

Linear Coefficients

```python
print(regressor.coef_)
```

```python
scores = cross_val_score(regressor, X_train, y_train, cv=10)
print(scores)

# Make cross validated predictions
predictions = cross_val_predict(regressor, X_train, y_train, cv=10)
plt.figure(figsize=(10,8))
sns.regplot(x=np.exp(y_train), y=np.exp(predictions))
plt.xlabel('Actual Value')
plt.ylabel('Prediction Value')

linear_cv_score = metrics.r2_score(y_train, predictions)
print("Linear Cross Validation Score", linear_cv_score)

linear_cv_rmse = np.sqrt(mean_squared_error(y_train, predictions))
print("RMSE :", linear_cv_rmse)
```

# Ridge Regression 

```python
ridge_alpha = 0
ridge_alphas = np.arange(0.001,1,0.001)

prev_ridge_score = 0

ridge_scores = []

for alpha in ridge_alphas:
    if alpha is 0:
        ridge_scores.append(0)
    if alpha is not 0:
        ridge_regressor = Ridge(alpha=alpha, normalize=True)
        ridge_regressor.fit(X_train, y_train)
        ridge_scores.append(ridge_regressor.score(X_test, y_test))
        if (prev_ridge_score < ridge_regressor.score(X_test, y_test)):
            ridge_alpha = alpha
            prev_ridge_score = ridge_regressor.score(X_test, y_test)

plt.figure(figsize=(10,8))
plt.scatter(ridge_alphas, ridge_scores)
```

```python
ridge_regressor = Ridge(alpha=ridge_alpha, normalize=True)
ridge_regressor.fit(X_train, y_train)

ridge_score = ridge_regressor.score(X_test, y_test)

print("Ridge Score : ", ridge_score)
print("Alpha Value : ", ridge_alpha)
ridge_predictions = ridge_regressor.predict(X_test)

plt.figure(figsize=(10,8))
# plt.plot(y_test, ridge_predictions, 'ro')
sns.regplot(x=np.exp(y_test), y=np.exp(ridge_predictions))
plt.xlabel('Actual Value')
plt.ylabel('Prediction Value')


ridge_rmse = np.sqrt(mean_squared_error(y_test, ridge_predictions))
print("RMSE :", ridge_rmse)

print("Ridge Intercept", ridge_regressor.intercept_)
```

Ridge Coefficients

```python
print(ridge_regressor.coef_)
```

```python
scores = cross_val_score(ridge_regressor, X_train, y_train, cv=10)

# Make cross validated predictions
predictions = cross_val_predict(ridge_regressor, X_train, y_train, cv=10)
plt.figure(figsize=(10,8))
# plt.scatter(y_train, predictions)
sns.regplot(x=np.exp(y_train), y=np.exp(predictions))
plt.xlabel('Actual Value')
plt.ylabel('Prediction Value')

ridge_cv_score = metrics.r2_score(y_train, predictions)
print("Ridge Cross Validation Score", ridge_cv_score)

# rmse = np.sqrt(mean_squared_error(y_train, predictions))
ridge_cv_rmse = np.sqrt(mean_squared_error(y_train, predictions))
print("RMSE :", ridge_cv_rmse)

print("Ridge Scores :", scores)
```

# Lasso Regression

Finding The Best Alpha Value For Lasso

```python
lasso_alpha = .0001
lasso_alphas = np.arange(0.00001,0.01,0.00001)
# lasso_alphas = np.arange(1,100,1)

prev_lasso_score = 0

lasso_scores = []

for alpha in lasso_alphas:
    if alpha is 0:
        lasso_scores.append(0)
    if alpha is not 0:
        lasso_regressor = Lasso(alpha=alpha, normalize=False, max_iter=1e5)
        lasso_regressor.fit(X_train, y_train)
        lasso_scores.append(lasso_regressor.score(X_test, y_test))
        if (prev_lasso_score < lasso_regressor.score(X_test, y_test)):
            lasso_alpha = alpha
            prev_lasso_score = lasso_regressor.score(X_test, y_test)
        
plt.figure(figsize=(10,8))
plt.scatter(lasso_alphas, lasso_scores)
```

Implementing Lasso Regression

```python
lasso_regressor = Lasso(alpha=lasso_alpha, normalize=False, max_iter=1e5)

lasso_regressor.fit(X_train, y_train)
lasso_score = lasso_regressor.score(X_test, y_test)

print("Lasso Score : ", lasso_score)
print("Alpha Value : ", lasso_alpha)
lasso_predictions = lasso_regressor.predict(X_test)

plt.figure(figsize=(10,8))
# plt.plot(y_test, lasso_predictions, 'ro')
sns.regplot(x=np.exp(y_test), y=np.exp(lasso_predictions))
plt.xlabel('Actual Value')
plt.ylabel('Prediction Value')
plt.show()

lasso_rmse = np.sqrt(mean_squared_error(y_test, lasso_predictions))
print("RMSE :", lasso_rmse)

print("Lasso Intercept", lasso_regressor.intercept_)
```

Lasso Coefficients

```python
print(lasso_regressor.coef_)
```

Lasso Cross Validation

```python
scores = cross_val_score(lasso_regressor, X_train, y_train, cv=10)

# Make cross validated predictions
predictions = cross_val_predict(lasso_regressor, X_train, y_train, cv=10)
plt.figure(figsize=(10,8))
# plt.scatter(y_train, predictions)
sns.regplot(x=np.exp(y_train), y=np.exp(predictions))
plt.xlabel('Actual Value')
plt.ylabel('Prediction Value')

lasso_cv_score = metrics.r2_score(y_train, predictions)
print("Lasso Cross Validation Score", lasso_cv_score)

lasso_cv_rmse = np.sqrt(mean_squared_error(y_train, predictions))
print("RMSE :", lasso_cv_rmse)
print("Lasso Scores : ", scores)
```
